## Matplotlib 사용

```python
# import 
import matplotlib.pyplot as plt
```

```python
x = [1,2,3]
y = [4,5,6]
```

```python
plt.figure()
plt.plot(x, y)
```

```python
# 선 두께
plt.plot(x, y, linewidth = 10)
```

```python
## 색상변경
plt.plot(x, y, linewidth = 10, color='red')
```

```python
# marker 각 데이터의 지정위치 표시 
plt.plot(x, y, marker='*') # *, o
```

```python
# marker size 변경
plt.plot(x, y, marker='*', markersize= 10) # *, o
```

```python
# 선 종류 선택
plt.plot(x, y, linestyle=':')
```

```python
# marker만 찍을때
plt.plot(x, y, 'o')
```

- subplot()
  - subplot(행의수 , 열의수,  해당 그래프가 그려질 위치)

```python
plt.subplot(1,2,1)
plt.plot(x1, y1)
plt.title('data1')

plt.subplot(1,2,2)
plt.plot(x1, y2)
plt.title('data2')
```

```python
plt.subplot(2,1,1)
plt.plot(x1, y1)
plt.title('data1')

plt.subplot(2,1,2)
plt.plot(x1, y2)
plt.title('data2')
```

```python
# 축이 2개인 경우의 그래프
x1 = [1,2,3]
y1 = [1,2,3]
y2 = [1, 50, 200]
fig, axe1 = plt.subplots()

axe2 = axe1.twinx()
chart1 = axe1.plot(x1, y1, color='red', label='data1')
chart2 = axe2.plot(x1, y2, color='blue', label='data2')

axe1.set_xlabel('x', fontsize=10)
axe1.set_ylabel('y1', fontsize=10)
axe2.set_ylabel('y2', fontsize=10)

# 범례추가
chart = chart1+ chart2
axe1.legend(chart,['data1','data2'])
```

- 가로로 그리기

```python
chart = sns.barplot(data= df, x='viewer', y = 'movietitle',
                    order = df.sort_values('viewer').movietitle)

xlabels = ['{:,.0f}'.format(i)+ 'k' for i in chart.get_xticks()/ 1000]
chart.set_xticklabels(xlabels)
```

```python
# bar 색상변경
chart = sns.barplot(data= df, x='viewer', y = 'movietitle',
                    order = df.sort_values('viewer').movietitle,
                    color='blue')

xlabels = ['{:,.0f}'.format(i)+ 'k' for i in chart.get_xticks()/ 1000]
chart.set_xticklabels(xlabels)
# 제목 및 라벨 추가/폰트 변경
plt.xlabel('viewer', fontsize = 15)
plt.ylabel('title', fontsize = 15)
plt.title('movie per viewer', fontsize = 20)
```

- pie 그리기

```python
movie_title= ['korea','it','academy']
audience = [664308, 2099131, 20067]
data = {'movietitle':movie_title,'viewer':audience}
df = pd.DataFrame(data)
```

```python
df.head()
```

```python
plt.pie(df['viewer'])
```

```python
# label 정보 기입
plt.pie(df['viewer'], labels=df['movietitle'])
```

```python
# 값 추가
plt.pie(df['viewer'], labels=df['movietitle'], autopct = '%0.2f%%')
```

```python
# color custom
# color_list = ['red','green','blue']
color_list = ['#FF0000','#A0522D','blue']
plt.pie(df['viewer'], labels=df['movietitle'], autopct = '%0.2f%%', colors=color_list)
```

```python
# explode를 이용해서 중심으로 부터 떼어내기
explode_list= [0.1, 0.1, 0.1]
plt.pie(df['viewer'], labels=df['movietitle'], autopct = '%0.2f%%', explode= explode_list)
```
